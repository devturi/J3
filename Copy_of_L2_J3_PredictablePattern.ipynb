Here, we will create a class of vehicle.. which will contain the required attributes of a vehicle like reputation, location, PID, RID etc


- Our purpose here is to create a vehicle reputation, that will have some kind of pattern in it. Because we know that vehicles reputation either grows or decreases by a pre-defined number. It can be either 0.2 or 0.3 etc


- The PID will change after every message as discussed in many papers

- We will create different instances of the vehicle class and then create reputation for each one of them.

- After the repuation of each vehicles is calculated, we will add it all to the dataframe, along with their changing PIDs. 

```python
class Vehicle:


  locationsArray = ['TownA','TownB','TownC','TownD','TownE','TownF']
  endLimit = len(locationsArray)-1

  def __init__(self):
        eth_k = generate_eth_key()
        sk_hex = eth_k.to_hex()
        pk_hex = eth_k.public_key.to_hex() 
        self.RID = pk_hex
        self.PID = 0
        self.reputation = 0.5
        self.currentLocation = 'ini'
        self.messageTimestamp = 0
        self.locationsArray = ['TownA','TownB','TownC','TownD','TownE','TownF']

       # self.previousLocation = 'ini'

  def changePID(self):
    eth_k = generate_eth_key()
    sk_hex = eth_k.to_hex()
    pk_hex = eth_k.public_key.to_hex()
    self.PID = pk_hex

  def randomLocation(self):
    if self.currentLocation =='ini':
      self.currentLocation = self.locationsArray[random.randint(0,5)]
    else:
      prevLocationIndex = self.locationsArray.index(self.currentLocation)
      if prevLocationIndex+1 >= len(self.locationsArray):
        self.currentLocation = self.locationsArray[0]
      else:
        self.currentLocation = self.locationsArray[prevLocationIndex+1]

  def sendMesg(self):
    self.messageTimestamp = time.time_ns()
    mesgType = ['true','false']
    mesg = mesgType[random.randint(0,1)]
    if mesg == 'true':
      self.changePID()
      if self.reputation+0.02 >=1:
        pass
      else:
        self.reputation = self.reputation +0.02
      self.randomLocation()
      
      
    else:
      self.changePID()
      if self.reputation-0.02 <= 0:
        pass
      else:
        self.reputation = self.reputation -0.02
      self.randomLocation()
      



v1 = Vehicle()
```

```python
def generate_like_df(df, categorical_cols, continuous_cols, best_distributions, n, seed=0):
    np.random.seed(seed)
    d = {}

    for c in categorical_cols:
        counts = df[c].value_counts()
        d[c] = np.random.choice(list(counts.index), p=(counts/len(df)).values, size=n)

    for c, bd in zip(continuous_cols, best_distributions):
        dist = getattr(scipy.stats, bd[0])
        d[c] = dist.rvs(size=n, *bd[1])

    return pd.DataFrame(d, columns=categorical_cols+continuous_cols)
```

```python
def isKAnonymized(df, k):
    for index, row in df.iterrows():
        query = ' & '.join([f'{col} == "{row[col]}"' for col in df.columns])
        rows = df.query(query)
        if rows.shape[0] < k:
            return False
    return True
```
